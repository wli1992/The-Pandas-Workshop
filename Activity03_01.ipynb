```python
#
# libraries
#
import pandas as pd
import sqlite3
```

```python
#
# show all tables in supply_company.db
#
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type = 'table'",
                     sqlite3.connect('../datasets/supply_company.db'))
tables
```

```python
#
# now read the entire Customers list
#
orders = pd.read_sql("select * from Orders",
                        sqlite3.connect("../datasets/supply_company.db"))
orders
```

```python
#
# total the orders by customer and find the largest
#
largest_cust = \
    orders.groupby('Customer_Number').sum()['amount'].\
                    sort_values(ascending = False).index[0]
largest_cust
```

```python
#
# now retrieve the corresponding customer record
#
largest_cust_info = \
    pd.read_sql(("select * from Customers WHERE Customer_Number = " +
                 str(largest_cust)),
                sqlite3.connect("../datasets/supply_company.db"))
largest_cust_info
```
