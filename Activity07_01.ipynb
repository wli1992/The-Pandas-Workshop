```python
#2 Import the pandas package:
import pandas as pd
```

```python
#3 Load the CSV file as a DataFrame:
file_url = 'https://raw.githubusercontent.com/PacktWorkshops/The-Pandas-Workshop/master/Chapter06/Data/student-mat.csv'
data_frame = pd.read_csv(file_url, delimiter=';')
```

```python
#4 Modify the DataFrame to contain only these columns:
#  “school”, “sex”, “age”, “address”, “heath”, “absences”, “G1”, “G2”, “G3”
data_frame = data_frame[[
    'school', 'sex', 'age', 'address', 'health', 'absences', 'G1', 'G2', 'G3'
]]
```

```python
#5 Display the first 10 rows of the DataFrame:
data_frame.head(10)
```

```python
#6 Build a pivot table indexed on "school":
pd.pivot_table(data_frame, index=['school'])
```

```python
#7 Build a pivot table indexed on "school" and "age":
pd.pivot_table(data_frame, index=['school', 'age'])
```

```python
#8 Build a pivot table indexed on "school", "sex" and "age"
#  with the mean and sum aggregation on the "absences" column:
pd.pivot_table(data_frame, index=['school', 'sex', 'age'], values = 'absences', aggfunc={'mean', 'sum'})
```
